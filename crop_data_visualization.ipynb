```python
# Import libraris

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# create a DataFrame from csv file
df = pd.read_csv('/content/Crop_recommendation.csv')
```

```python
# Print the head of df
print(df.head())
```

```python
df.describe()
```

```python
df.shape
```

```python
# Return the count of each unique label
df['label'].value_counts()
```

```python
# Plotting a pairplot to visualize relationships between variables
sns.pairplot(df, hue='label')
plt.show()
```

```python
# Boxplot to visualize distribution of temperature across different crops
plt.figure(figsize=(10, 6))
sns.boxplot(x='label', y='temperature', data=df)
plt.title('Temperature Distribution Across Crops')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot to visualize the mean nitrogen levels for each crop
plt.figure(figsize=(10, 6))
sns.barplot(x='label', y='N', data=df, errorbar=None)
plt.title('Mean Nitrogen Levels Across Crops')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average level of nutirients by crop
nutrients = ['N', 'P', 'K']
plt.figure(figsize=(12, 6))
for nutrient in nutrients:
    sns.barplot(x='label', y=nutrient, data=df,errorbar=None)
plt.title('Average Nutrient Levels by Crop')
plt.xlabel('Crop')
plt.ylabel('Average Level')
plt.xticks(rotation=45)
plt.legend(nutrients)
plt.show()
```

```python
# Relationship between Rainfall and Crop Yield
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rainfall', y='humidity', hue='label', data=df)
plt.title('Relationship between Rainfall and Crop Yield')
plt.xlabel('Rainfall')
plt.ylabel('Ph')
plt.legend(title='Crop')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temperature', y='humidity', hue='label', data=df)
plt.title('Temperature vs. Humidity by Crop')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.legend(title='Crop')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='label', y='ph', data=df)
plt.title('pH Distribution Across Crops')
plt.xlabel('Crop')
plt.ylabel('pH')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

# Assuming temperatures_ind is your DataFrame with country and city in the index

# Sorting the index of temperatures_ind
temperatures_srt = temperatures_ind.sort_index()

# Slicing to get subsets
# Subset from Pakistan to Russia
print(temperatures_srt.loc['Pakistan':'Russia'])

# Subset from Lahore to Moscow
print(temperatures_srt.loc['Lahore':'Moscow'])

# Subset from Pakistan, Lahore to Russia, Moscow
print(temperatures_srt.loc[('Pakistan', 'Lahore'):('Russia', 'Moscow')])
```

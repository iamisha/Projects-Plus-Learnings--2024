* Create a displot() of the fmr_2 column in df using a dark style. Use plt.clf() to clear the figure.

```python
sns.set_style("dark")
sns.displot(df["fmr_2"])
plt.clf()
plt.show()
```

* Create the same displot() of fmr_2 using a whitegrid style. Clear the plot after showing it.

```python
sns.set_style("whitegrid")
sns.displot(df["fmr_2"])
plt.show()
plt.clf()
```

**Exercise 2**

```python
# Set the style to white
sns.set_style('white')

# Create a regression plot
sns.lmplot(data=df,
           x='pop2010',
           y='fmr_2')

# Remove the spines
sns.despine(right=True, top=True)

# Show the plot and clear the figure
plt.show()
plt.clf()
```

* In a plot, the term "spines" refers to the lines that border the axes on all sides, forming the box-like frame around the plot. By removing the spines, you eliminate these lines, resulting in a cleaner and less cluttered appearance for the plot.

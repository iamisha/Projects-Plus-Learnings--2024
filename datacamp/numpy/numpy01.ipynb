**The .dtype attribute**

* We can find the data type of elements in an array using the dot-dtype array atribute

```python
np.array([1.32, 5,78, 175.55]).dtype
```

**Default data types**

```python
int_array = np.array([[1,2,3], [4,5,6]])
int_array.dtype
```

**String data**

```python
np.array(["Introduction", "to", "NumPy"]).dtype
```

**dtype as an argument**

```python
float32_array = np.array([1.32, 5,78, 175.55], dtype = np.float32)
float32_array.dtype
```

```python
int32_array = np.array([1, 5, 175])
int32_array.dtype
```

```python
int_array = np.array([1, 5, 175], dtype = np.float32)
int_array.dtype
```

**Type Conversion**

```python
boolean_array = np.array([[True, False], [False, True]])
print(boolean_array.dtype)
boolean_array.astype(np.float32)
```

**type coercion:** Type coercion is the process of automatic or implicit conversion of values from one data type to another.

```python
np.array([True, "boop", 42, 42.43])
```

**Type coercion hierarchy**

Adding a float to an array of integers will change all integers into float

```python
np.array([0,42, 42.42]).dtype
```

* Adding an integers to an array of booleans will change all booleans in to integers

```python
np.array([True, False, 42]).dtype
```

**Exercise**: Converting the datatype of the zeros array to the int32 from float64



```python
# Create an array of zeros with three rows and two columns
zero_array = np.zeros((3, 2))

# Print the data type of zero_array
print(zero_array.dtype)

# Create a new array of int32 zeros with three rows and two columns
zero_int_array = np.zeros((3,2), dtype = np.int32)

# Print the data type of zero_int_array
print(zero_int_array.dtype)
```

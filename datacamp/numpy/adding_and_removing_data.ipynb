```python
import numpy as np
import matplotlib.pyplot as plt
```

![concatenating.png](shape_compatibility.png)

**Concatenating rows**

```python
classroom_ids_and_sizes = np.array([[1,22], [2,21], [3,27], [4,26]])
new_classrooms = np.array([[5,30], [5,17]])
np.concatenate((classroom_ids_and_sizes, new_classrooms))
```

* np.concatenate() concatenates along the first axis by default

**Concatenating columns**

```python
classroom_ids_and_sizes = np.array([[1,22], [2,21], [3,27], [4,26]])
grade_levels_and_teachers = np.array([[1,"James"], [1, "George"], [3, "Ammy"], [3, "Meehir"]])
classroom_data = np.concatenate((classroom_ids_and_sizes, grade_levels_and_teachers), axis = 1)
print(classroom_data)
```

**Creating Compatibility**

```python
array_1D = np.array([1,2,3])
column_array_2D = array_1D.reshape((3,1))
column_array_2D
```

**Deleting with np.delete()**

* To delete row, axis = 0

```python
np.delete(classroom_data, 2, axis=0)
```

* To delete column, set axis = 1

```python
np.delete(classroom_data, 0, axis = 1)
```

**Aggregating methods**



*   .sum()
*   .min()
*   .max()
*   .mean()
*   .cumsum()



```python
security_breaches = np.zeros((5, 3), dtype = np.int32)
security_breaches

security_breaches[0,1] = 5
security_breaches[0,2] = 1
security_breaches[1,1] = 2
security_breaches[2,0] = 1
security_breaches[2,1] = 1
security_breaches[2,2] = 2
security_breaches[3,0] = 2
security_breaches[3,1] = 2
security_breaches[3,2] = 1

security_breaches
```

**Summing data**

.sum(): method adds up all elements in the entire array.

```python
security_breaches.sum()
```

**Aggregating rows**

We can control which axis to sum across with the axis argument

axis = 0 for row

```python
security_breaches.sum(axis=0)
```

**Aggregating columns**

setting axis = 1 will sum the values of all columns in each row, creating row totals

```python
security_breaches.sum(axis=1)
```

**Minimum and maximum values**

```python
security_breaches.min()
```

```python
security_breaches.max()
```

return the min or max of each column if the axis argument is set to 0

```python
security_breaches.min(axis=0)
```

return the min or max of each column if the axis arugment is set to 1

```python
security_breaches.max(axis=0)
```

```python
security_breaches.min(axis=1)
```

```python
security_breaches.max(axis=1)
```

**Finding the mean**

```python
security_breaches.mean()
```

calculate the mean of each row

```python
security_breaches.mean(axis=1)
```

calculate the mean of each column

```python
security_breaches.mean(axis=0)
```

**The Keepdims argument**

```python
security_breaches.sum(axis=1)
```

```python
security_breaches.sum(axis=1, keepdims=True)
```

```python
security_breaches.cumsum(axis = 0)
```

**Graphing summary values**

```python
cum_sums_by_client = security_breaches.cumsum(axis=0)
plt.plot(np.arange(1,6), cum_sums_by_client[:, 0], label = "Client 1")
plt.plot(np.arange(1,6), cum_sums_by_client.mean(axis=1), label = "Average")
plt.legend()
plt.show()
```

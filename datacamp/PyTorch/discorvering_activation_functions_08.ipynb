##Sigmoid Function:
#### Limitation:

Gets Confused Easily: Imagine you have a toy that can only say "yes" or "no" when you ask it questions. The sigmoid function is like that toy. But sometimes, when the question is too tricky, the toy can't decide properly and gives an answer that's not very clear. So, if you ask it if something is "a little bit yes and a little bit no," it gets really confused and might not help you much.


##Softmax Function:
#### Limitation:

Only Likes One Best Answer: Now, think of a game where you have to pick your favorite ice cream flavor out of chocolate, vanilla, and strawberry. The softmax function is like a judge in this game who always wants to pick just one winner. But if you really like chocolate and vanilla almost equally, it will still make you pick just one as the best, even though you like both a lot. So, it sometimes ignores the fact that you might have more than one favorite.

## Implementing ReLU

```python
# Create a ReLU function with PyTorch
relu_pytorch = nn.ReLU()

# Apply your ReLU function on x, and calculate gradients
x = torch.tensor(-1.0, requires_grad=True)
y = relu_pytorch(x)
y.backward()

# Print the gradient of the ReLU function for x
gradient = x.grad
print(gradient)
```

## Implementing LeakyReLU

```python
# Create a leaky relu function in PyTorch
leaky_relu_pytorch = nn.LeakyReLU(negative_slope=0.05)

x = torch.tensor(-2.0)
# Call the above function on the tensor x
output = leaky_relu_pytorch(x)
print(output)
```

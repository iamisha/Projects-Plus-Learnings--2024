```python
import torch
import torch.nn as nn

class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.layer1 = nn.Linear(12, 64)  # First hidden layer
        self.layer2 = nn.Linear(64, 32)  # Second hidden layer
        self.layer3 = nn.Linear(32, 16)  # Third hidden layer
        self.output = nn.Linear(16, 2)   # Output layer

    def forward(self, x):
        x = torch.relu(self.layer1(x))  # Apply ReLU activation after first hidden layer
        x = torch.relu(self.layer2(x))  # Apply ReLU activation after second hidden layer
        x = torch.relu(self.layer3(x))  # Apply ReLU activation after third hidden layer
        x = self.output(x)              # No activation function for the output layer
        return x

# Create the input tensor
input_tensor = torch.Tensor(
    [[2, 3, 6, 7, 9, 3, 2, 1, 5, 3, 6, 9]]
)

# Initialize the network
model = NeuralNetwork()

# Perform a forward pass
output = model(input_tensor)

# Print the output
print(output)
```

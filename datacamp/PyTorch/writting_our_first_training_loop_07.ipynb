## Using the MSELoss

```python
import numpy as np
import torch
import torch.nn as nn

# Given values
y_hat = np.array([10])
y = np.array([1])

# Calculate the MSELoss using NumPy
mse_numpy = np.mean((y_hat - y) ** 2)
print(f"MSE calculated using NumPy: {mse_numpy}")

# Create the MSELoss function using PyTorch
criterion = nn.MSELoss()

# Convert y_hat and y to tensors and then to float data types
y_hat_tensor = torch.tensor(y_hat, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)

# Calculate the MSELoss using the created loss function
mse_pytorch = criterion(y_hat_tensor, y_tensor)
print(f"MSE calculated using PyTorch: {mse_pytorch.item()}")
```

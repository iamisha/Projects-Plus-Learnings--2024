**Exercise 1**

```python
class BankAccount:
    def __init__(self, number, balance):
        self.balance = balance
        self.number = number

    def withdraw(self, amount):
        self.balance -= amount

    # Define __eq__ that returns True if both number and balance attributes are equal
    def __eq__(self, other):
        return self.number == other.number and self.balance == other.balance

# Create accounts and compare them
acct1 = BankAccount(123, 1000)
acct2 = BankAccount(123, 1000)
acct3 = BankAccount(456, 1000)
print(acct1 == acct2)  # Should print True, as their numbers and balances are equal
print(acct1 == acct3)  # Should print False, as their numbers are equal but balances are different
print(acct1.number)
```
